```python
from ride.grapher import GraphProcessor, GraphRunner
from ride.utils import DataGetter
from ride.plotter import visualisation, box_visualisation
```

```python
G = DataGetter.download_graph(id="R44915")
```

```python
# Находим набор рабочих resolutions, ks, (v1, e1) для каждого графа
resolutions, ks, v1, e1 = GraphProcessor.search_resolutions(G, k_max=0.6)
```

```python
output, all_errors = GraphRunner.test(G, resolutions, k=5, weight='length')
```

```python
# Визуализация времени работы по ks и формуле
visualisation(G, v1, e1, True, False)    
box_visualisation(G, output['time_preprocess_dijkstra_on_centroid_len'], output['time_dijkstra_base'].iloc[0].item(), all_errors, True, False) 
```
