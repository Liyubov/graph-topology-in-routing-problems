```python
from gsber.grapher import GraphProcessor, GraphRunner
from gsber.utils import DataGetter
from gsber.plotter import visualisation, box_visualisation
```

```python
G = DataGetter.download_graph(id="R1430508")
```

```python
# Находим набор рабочих resolutions, ks, (v1, e1) для каждого графа
resolutions, ks, v1, e1 = GraphProcessor.search_resolutions(G, k_max=0.6)
output, mistakes = GraphRunner.test(G, resolutions, k=5, weight='length')

# Визуализация времени работы по ks и формуле
visualisation(G, v1, e1, True, False)    
box_visualisation(G, output, mistakes, True, False) 
```
